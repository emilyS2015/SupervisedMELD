```python
%matplotlib inline
from sklearn.cross_validation import KFold
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
def MELD_Backward(data, Numfolds):
    """This function performs supervised MELD and backward elimination algorithm
        Input: a dataframe and number of folds
        output: select maximum AUC and corresponding features 
        for every specific number of features  
    """
    #divide data into kf cross validation folds
    kf=KFold(data.shape[1],Numfolds,shuffle=True,random_state=111)
    
    def profile_class(data,phi):
        """This function calculates the GoM scores for each individual
        Input: data and phi calculated from MELD
           Output: GoM score
        """
        score=np.zeros([data.shape[1],2])
        profile=np.zeros([int(data.shape[0]),int(data.shape[1])])
        for j in range(data.shape[1]):
            for i in range(data.shape[0]):
                yij=int(data[i,j])
                #calculate M_ij
                profile[i,j]=np.argmax(phi[i][:,yij])
                if profile[i,j]==0:
                    score[j,0]=score[j,0]+1
                else:
                    score[j,1]=score[j,1]+1
            score[j,:]=(score[j,:]+0.1)/(phi.shape[0]+0.2)
        return score
    
    #create a list of features
    vec_list=[]
    for j in range(1,data.shape[0]):
        integers=list(range(0,data.shape[0]))
        integers.remove(j)
        vec_list.append(integers)
    
    auc_list_outer=[]    
    for j in range(len(vec_list)):
        auc_list=[]
        vec=vec_list[j]
        for i in range(Numfolds):
            train=data[:,list(kf)[i][0]]
            test=data[:,list(kf)[i][1]]
            import MELD_updated as MELD
            k=2
            S=100
            Y = train
            (p,n) = Y.shape
            Yt = np.array([0]*p)
            Yt[0]=0
            # create an object of MELD class
            myMELD = MELD.MELD(Y,Yt,k,vec)
            # calculate second moment matrices
            myMELD.calM2()
            myMELD.calM2_bar()
            # ------------- first stage
            # initialize weight matrices to identity
            myMELD.initializeWeight_M2()
            # perform first stage estimation
            Result1 = myMELD.estimatePhiGrad_M2(S)
            # extract phi from the result
            phi1=Result1['PHI'][Result1['iter']]
            # extract selected features in the test fold
            test=test[vec]
            # calculate GoM score
            gscore=profile_class(test[1:],phi1[1:])
            # calculate risk
            risk=gscore[:,None]*np.transpose(phi1[0])
            risk=np.sum(risk[:,1,:],axis=1)
            # replace 0 with label "-1"
            np.place(test[0],test[0]==0,[-1])
            # calculate AUC
            auc = roc_auc_score(test[0],risk)
            auc_list.append(auc)
        #calculate the mean AUC across kf folds
        auc_new=np.mean(auc_list)
        auc_list_outer.append(auc_new)
        
    auc_all=[]
    vec_all=[]

    while len(vec_list)>2:
        vec_list_2=vec_list[np.argmax(auc_list_outer).astype(int)]
        vec_all.append(vec_list_2)
        vec_list=[]
        #eliminate one feature at a time
        for k in range(len(vec_list_2)-1):
            integers=vec_list_2[:]
            del integers[k+1]
            vec_list.append(integers)

        auc_list_outer=[] 

        for j in range(len(vec_list)):
            auc_list=[]
            vec=vec_list[j]

            for i in range(10):
            
                train=data[:,list(kf)[i][0]]
    
                test=data[:,list(kf)[i][1]]
        
                import MELD_updated as MELD
                k=2
                S=100

                Y = train
                (p,n) = Y.shape

                Yt = np.array([0]*p)
                Yt[0]=0
                # create an object of MELD class
                myMELD = MELD.MELD(Y,Yt,k,vec)

                # calculate second moment matrices
                myMELD.calM2()
                myMELD.calM2_bar()

                # ------------- first stage
                # initialize weight matrices to identity
                myMELD.initializeWeight_M2()

                # perform first stage estimation
                Result1 = myMELD.estimatePhiGrad_M2(S)
                # extract phi from the result
                phi1=Result1['PHI'][Result1['iter']]
                # extract selected features from test fold
                test=test[vec]
                # calculate GoM score
                gscore=profile_class(test[1:],phi1[1:])
                # calculate risk
                risk=gscore[:,None]*np.transpose(phi1[0])
                risk=np.sum(risk[:,1,:],axis=1)
                # replace 0 with label "-1"
                np.place(test[0],test[0]==0,[-1])
                # calculate AUC
                auc = roc_auc_score(test[0],risk)
                auc_list.append(auc)
            # calculate average AUC across kf folds
            auc_new=np.mean(auc_list)
            
            auc_list_outer.append(auc_new)
        # selected the maximum AUC from a list of AUC's
        auc_new_new=max(auc_list_outer)
    
        auc_all.append(auc_new_new)
    
    return auc_all, vec_all


```

```python
def Supervised_MELD_Features(data, Numfolds, features):
    """This function gives auc scores given data, number of folds and the features
    """
    
    def profile_class(data,phi):
        """This function calculates the GoM scores for each individual
        Parameters:
        Input: data and phi calculate from MELD
           Output: GoM score
        """
        score=np.zeros([data.shape[1],2])
        profile=np.zeros([int(data.shape[0]),int(data.shape[1])])
        for j in range(data.shape[1]):
            for i in range(data.shape[0]):
                yij=int(data[i,j])
                profile[i,j]=np.argmax(phi[i][:,yij])
                if profile[i,j]==0:
                    score[j,0]=score[j,0]+1
                else:
                    score[j,1]=score[j,1]+1
            score[j,:]=(score[j,:]+0.1)/(phi.shape[0]+0.2)
        return score
    #divide data into kf folds
    kf=KFold(data.shape[1],Numfolds,shuffle=True,random_state=111)

    auc_list=[]
    
    fig = plt.figure(figsize=[12,8])
    for i in range(Numfolds):
        train=data[:,list(kf)[i][0]]
    
        test=data[:,list(kf)[i][1]]
    
        vec=features


        import MELD_85 as MELD
        k=2
        S=100

        Y = train
        (p,n) = Y.shape

        Yt = np.array([0]*p)
        Yt[0]=0
        # create an object of MELD class
        myMELD = MELD.MELD(Y,Yt,k,vec)

        # calculate second moment matrices
        myMELD.calM2()
        myMELD.calM2_bar()

        # ------------- first stage
        # initialize weight matrices to identity
        myMELD.initializeWeight_M2()

        # perform first stage estimation
        Result1 = myMELD.estimatePhiGrad_M2(S)
        # extract phi from the MELD result
        phi1=Result1['PHI'][Result1['iter']]
        # extract selected features from the test fold
        test=test[vec]
        # replace 0 with label "-1" in the test fold
        np.place(test[0],test[0]==0,[-1])
        # calculate GoM score
        gscore=profile_class(test[1:],phi1[1:])
        
        risk=gscore[:,None]*np.transpose(phi1[0])
    
        risk=np.sum(risk[:,1,:],axis=1)
    
        fpr,tpr,thresh= roc_curve(test[0],risk)
    
        auc= roc_auc_score(test[0],risk)
        
        axis=fig.add_subplot(2,5,i+1)
    
        axis.plot(fpr,tpr)
    
        axis.plot([0,1],[0,1],"r--",alpha=.5)
        
        axis.set_xlabel("FPR")
        
        axis.set_ylabel("TPR")
    
        print("AUC: {}".format(auc))
    
        auc_list.append(auc)  
    
    fig.suptitle('ROC Curve (After Supervised MELD & Backward Elimination)',fontsize=20)
    
    fig.tight_layout()
    
    fig.subplots_adjust(top=0.9)
    
    plt.show()
    
    fig.savefig('ROC_Sup_Bac.pdf')
        
    return auc_list
```

```python
def Supervised_MELD(data, Numfolds, features):
    """This function calculates the auc and risks
    """
    
    def profile_class(data,phi):
        """This function calculates the GoM scores for each individual
        Parameters:
        Input: data and phi calculate from MELD
           Output: GoM score
        """
        score=np.zeros([data.shape[1],2])
        profile=np.zeros([int(data.shape[0]),int(data.shape[1])])
        for j in range(data.shape[1]):
            for i in range(data.shape[0]):
                yij=int(data[i,j])
                profile[i,j]=np.argmax(phi[i][:,yij])
                if profile[i,j]==0:
                    score[j,0]=score[j,0]+1
                else:
                    score[j,1]=score[j,1]+1
            score[j,:]=(score[j,:]+0.1)/(phi.shape[0]+0.2)
        return score
    #divide data into kf folds
    kf=KFold(data.shape[1],Numfolds,shuffle=False,random_state=111)

    auc_list=[]
    
    fig = plt.figure(figsize=[12,8])
    risk_list=[]
    for i in range(Numfolds):
        train=data[:,list(kf)[i][0]]
    
        test=data[:,list(kf)[i][1]]
    
        vec=features


        import MELD_85 as MELD
        k=2
        S=100

        Y = train
        (p,n) = Y.shape

        Yt = np.array([0]*p)
        Yt[0]=0
        # create an object of MELD class
        myMELD = MELD.MELD(Y,Yt,k,vec)

        # calculate second moment matrices
        myMELD.calM2()
        myMELD.calM2_bar()

        # ------------- first stage
        # initialize weight matrices to identity
        myMELD.initializeWeight_M2()

        # perform first stage estimation
        Result1 = myMELD.estimatePhiGrad_M2(S)
    
        phi1=Result1['PHI'][Result1['iter']]
    
        test=test[vec]
    
        np.place(test[0],test[0]==0,[-1])
    
        gscore=profile_class(test[1:],phi1[1:])
    
        risk=gscore[:,None]*np.transpose(phi1[0])
    
        risk=np.sum(risk[:,1,:],axis=1)
        
        risk_list.append(risk)
    
        fpr,tpr,thresh= roc_curve(test[0],risk)
    
        auc= roc_auc_score(test[0],risk)
        
        axis=fig.add_subplot(2,5,i+1)
    
        axis.plot(fpr,tpr)
    
        axis.plot([0,1],[0,1],"r--",alpha=.5)
        
        axis.set_xlabel("FPR")
        
        axis.set_ylabel("TPR")
    
        print("AUC: {}".format(auc))
    
        auc_list.append(auc)
    
    
    fig.suptitle('ROC Curve (Supervised MELD)',fontsize=16)
    
    fig.tight_layout()
    
    fig.subplots_adjust(top=0.90)
    
    plt.show()
    
    fig.savefig('ROC_Sup.pdf')
    
    risk_list=np.concatenate(risk_list)
        
    return auc_list, risk_list
```

```python
def MELD_Backward_features(data, Numfolds, features):
    """This function performs supervised MELD and backward elimination algorithm
        Input: a dataframe and number of folds
        output: select maximum AUC and corresponding features 
        for every specific number of features  
    """
    #divide data into kf cross validation folds
    kf=KFold(data.shape[1],Numfolds,shuffle=False,random_state=111)
    
    def profile_class(data,phi):
        """This function calculates the GoM scores for each individual
        Parameters:
        Input: data and phi calculate from MELD
           Output: GoM score
        """
        score=np.zeros([data.shape[1],2])
        profile=np.zeros([int(data.shape[0]),int(data.shape[1])])
        for j in range(data.shape[1]):
            for i in range(data.shape[0]):
                yij=int(data[i,j])
                #calculate M_ij
                profile[i,j]=np.argmax(phi[i][:,yij])
                if profile[i,j]==0:
                    score[j,0]=score[j,0]+1
                else:
                    score[j,1]=score[j,1]+1
            score[j,:]=(score[j,:]+0.1)/(phi.shape[0]+0.2)
        return score
    
    #create a list of features
    vec_list=[]
    for j in range(len(features)-1):
        integers=features[:]
        del integers[j+1]
        vec_list.append(integers)
    
    auc_list_outer=[]
    whole_gom_list=[]
    for j in range(len(vec_list)):
        auc_list=[]
        gom_list=[]
        vec=vec_list[j]
        for i in range(Numfolds):
            train=data[:,list(kf)[i][0]]
            test=data[:,list(kf)[i][1]]
            import MELD_85 as MELD
            k=2
            S=100
            Y = train
            (p,n) = Y.shape
            Yt = np.array([0]*p)
            Yt[0]=0
            # create an object of MELD class
            myMELD = MELD.MELD(Y,Yt,k,vec)
            # calculate second moment matrices
            myMELD.calM2()
            myMELD.calM2_bar()
            # ------------- first stage
            # initialize weight matrices to identity
            myMELD.initializeWeight_M2()
            # perform first stage estimation
            Result1 = myMELD.estimatePhiGrad_M2(S)
            # extract phi from the result
            phi1=Result1['PHI'][Result1['iter']]
            # extract selected features in the test fold
            test=test[vec]
            # calculate GoM score
            gscore=profile_class(test[1:],phi1[1:])
            # calculate risk
            risk=gscore[:,None]*np.transpose(phi1[0])
            risk=np.sum(risk[:,1,:],axis=1)
            gom_list.append(gscore[:,1])
            # replace 0 with label "-1"
            np.place(test[0],test[0]==0,[-1])
            # calculate AUC
            auc = roc_auc_score(test[0],risk)
            auc_list.append(auc)
        #calculate the mean AUC across kf folds
        auc_new=np.mean(auc_list)
        auc_list_outer.append(auc_new)
        whole_gom_list.append(list(np.concatenate(gom_list)))
        
    auc_all=[]
    vec_all=[]
    gom_all=[]
    while len(vec_list)>2:
        vec_list_2=vec_list[np.argmax(auc_list_outer).astype(int)] 
        gom_list_2=whole_gom_list[np.argmax(auc_list_outer).astype(int)]
        vec_all.append(vec_list_2)
        gom_all.append(gom_list_2)
        vec_list=[]
        whole_gom_list=[]
        #eliminate one feature at a time
        for k in range(len(vec_list_2)-1):
            integers=vec_list_2[:]
            del integers[k+1]
            vec_list.append(integers)

        auc_list_outer=[] 
       
        for j in range(len(vec_list)):
            auc_list=[]
            gom_list=[]
            vec=vec_list[j]

            for i in range(10):
            
                train=data[:,list(kf)[i][0]]
    
                test=data[:,list(kf)[i][1]]
        
                import MELD_85 as MELD
                k=2
                S=100

                Y = train
                (p,n) = Y.shape

                Yt = np.array([0]*p)
                Yt[0]=0
                # create an object of MELD class
                myMELD = MELD.MELD(Y,Yt,k,vec)

                # calculate second moment matrices
                myMELD.calM2()
                myMELD.calM2_bar()

                # ------------- first stage
                # initialize weight matrices to identity
                myMELD.initializeWeight_M2()

                # perform first stage estimation
                Result1 = myMELD.estimatePhiGrad_M2(S)
                # extract phi from the result
                phi1=Result1['PHI'][Result1['iter']]
                # extract selected features from test fold
                test=test[vec]
                # calculate GoM score
                gscore=profile_class(test[1:],phi1[1:])
                # calculate risk
                risk=gscore[:,None]*np.transpose(phi1[0])
                risk=np.sum(risk[:,1,:],axis=1)
                gom_list.append(gscore[:,1])
                # replace 0 with label "-1"
                np.place(test[0],test[0]==0,[-1])
                # calculate AUC
                auc = roc_auc_score(test[0],risk)
                auc_list.append(auc)
            # calculate average AUC across kf folds
            auc_new=np.mean(auc_list)     
            auc_list_outer.append(auc_new)
            whole_gom_list.append(list(np.concatenate(gom_list)))
              
        
        # selected the maximum AUC from a list of AUC's
        auc_new_new=max(auc_list_outer)
    
        auc_all.append(auc_new_new)
        
        auc=np.max(auc_all)
        
        feature_selected=vec_all[np.argmax(auc_all)]
        
        gom=gom_all[np.argmax(auc_all)]
    
    return auc, feature_selected, gom


```

```python
def plot_GOM(gom_envir, gom_behav):
    """This function plot environmental GoM vs. behavioral GoM
    """
    fig = plt.figure(figsize=[8,5])
    #Generate a list of unique points
    points=list(set(zip(gom_behav,gom_envir))) 
    #Generate a list of point counts
    count=[len([x for x,y in zip(gom_behav,gom_envir) if x==p[0] and y==p[1]]) for p in points]
    #Now for the plotting:
    plot_x=[i[0] for i in points]
    plot_y=[i[1] for i in points]
    count=np.array(count)
    plt.scatter(plot_x,plot_y,c=count,s=100*count**0.5,cmap='Spectral_r')
    plt.xlabel('Behavioral GoM Score')
    plt.ylabel('Environmental GoM Score')
    cbar=plt.colorbar()
    cbar.set_label('Counts')
    fig.suptitle('GoM Scores for the Machadinho data in 1985',fontsize=16)
    plt.show()
    fig.savefig('GOM85.pdf')
```
